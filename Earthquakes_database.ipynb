```python
import pandas as pd
```

```python
df = pd.read_csv("Earthquakes_database.csv")
```

```python
df
```

## Identifies and removes duplicate rows within a dataset.

```python
df = df.drop_duplicates()
df
```

# Targets and removes the 'ID' row specifically

```python
df = df.drop(columns = "ID")
df
```

# splits a date column  in a DataFrame into its day, month, and year components

```python
df[['Day', 'Month', 'Year']] = df['Date'].str.split('/', expand=True)
df
```

# Fill all NaN values in the DataFrame  with empty spaces

```python
df=df.fillna('')
df
```

# Categorize magnitude values ranging from 5.5 to 9.1 into three distinct labels: 'Low', 'Medium', and 'High'.

```python
bins = [5.4, 7, 8, 9.2]  
labels = ['Low', 'Medium', 'High']  

df['Magnitude_Category'] = pd.cut(df['Magnitude'], bins=bins, labels=labels)

df
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_short = df.tail(50)
```

# Line Chart

```python
plt.figure(figsize=(8, 5))
plt.plot(df_short['Date'], df_short['Magnitude'], marker='o')
plt.title('Magnitude of Earthquakes over Dates')
plt.xlabel('Date')
plt.ylabel('Magnitude')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Bar Chart

```python
plt.figure(figsize=(8, 5))
plt.bar(df_short['Date'], df_short['Depth'], color='orange')
plt.title('Average Depth per Earthquake')
plt.xlabel('Date')
plt.ylabel('Depth')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

# Scatter Chart

```python
plt.figure(figsize=(8, 5))
plt.scatter(df_short['Magnitude'], df_short['Depth'], color='green', alpha=0.7)
plt.title('Magnitude vs Depth')
plt.xlabel('Magnitude')
plt.ylabel('Depth')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Pie Chart

```python
bins = [0, 5, 6, 7, np.inf]
labels = ['<5', '5-6', '6-7', '7+']

df_short['Magnitude_Category'] = pd.cut(df_short['Magnitude'], bins=bins, labels=labels)


magnitude_counts = df_short['Magnitude_Category'].value_counts()

plt.figure(figsize=(8, 5))
plt.pie(magnitude_counts, labels=magnitude_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Magnitudes')
plt.axis('equal')
plt.tight_layout()
plt.show()
```
